# Py City Schools Analysis
1. Comparing performances vs size of schools, the schools with < 2000 students(small and medium sizes)(See table: Scores by School) are the ones with better performances. So the fewer students a school has the better their performance is.
2.  Charter vs District schools: Charter schools tend to perform much better than the district schools. In the tables "Top Performing Schools" and " Botton Performing Schools" is clear that the charter school are the best schools and the district schools are the worst. It can be hypothesized that due to the size of charter schools, they performed better than district schools, that has more students.

```python
import pandas as pd

school_data_to_load = "Resources/schools_complete.csv"
student_data_to_load = "Resources/students_complete.csv"

school_data = pd.read_csv(school_data_to_load)
student_data = pd.read_csv(student_data_to_load)

school_data.head()
```

```python
student_data.head()
```

```python
school_data_complete = pd.merge(student_data, school_data, how="left", on=["school_name", "school_name"])
school_data_complete.head()
```

```python
total_schools = len(school_data_complete["school_name"].unique())
total_schools
```

```python
total_students = (school_data_complete["student_name"]).count()
total_students
```

```python
total_budget = (school_data_complete["budget"].unique()).sum()
total_budget 
```

```python
average_math = round(school_data_complete["math_score"].mean(),3)
average_math
```

```python
average_reading = round(school_data_complete["reading_score"].mean(),3)
average_reading
```

```python
passing_rate = round((school_data_complete["math_score"].mean() + school_data_complete["reading_score"].mean())/2,3)
passing_rate
```

```python
passing_math = school_data_complete.loc[school_data_complete["math_score"] >= 70]
passing_math.head()
```

```python
pas_math = round((((passing_math["math_score"].count())*100)/(school_data_complete["student_name"].count)()),3)
pas_math
```

```python
passing_reading = school_data_complete.loc[school_data_complete["reading_score"] >= 70]
passing_reading.head()
```

```python
pas_reading = round((((passing_reading["reading_score"].count())*100)/(school_data_complete["student_name"].count())),3)
pas_reading
```

# District Summary

```python
summary_district = pd.DataFrame({"Total Schools": [total_schools], "Total Students": [total_students], "Total Budget": [total_budget],"Average Math Score":[average_math], "Average Reading Score": [average_reading], "% Passing Math": [pas_math],"% Passing Reading": [pas_reading],"%Overall Passing Rate": [passing_rate]})
summary_district
```

```python
group_math = pass_math.groupby(['school_name'])
group_math["student_name"].count()
```

```python
p_math = round(((group_math["student_name"].count())*100)/(tot_student),3)
p_math = pd.DataFrame(p_math)
p_math = p_math.reset_index()
p_math
```

```python
group_reading = pass_reading.groupby(['school_name'])
group_reading["student_name"].count()
```

```python
p_reading = round(((group_reading["student_name"].count())*100)/(tot_student),3)
p_reading = pd.DataFrame(p_reading)
p_reading = p_reading.reset_index()
p_reading
```

```python
school_4 = pd.merge(school_3, p_reading, how="left", on=["school_name"])
school_4
```

```python
summary_schools = school_4.set_index("school_name")
summary_schools
```
